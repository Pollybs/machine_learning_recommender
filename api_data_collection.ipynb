```python
import requests as requests

import pandas as pd
```

# apiurl = https://unstats.un.org/sdgapi/
Endpoints: 

Goal List : GET /v1/sdg/Goal/List Returns the list of all Goals available


GeoArea List: GET /v1/sdg/GeoArea/List  Returns a flat list with all the geographies; geoAreaCode is the M49 Code

GET /v1/sdg/Indicator/List Returns the list of all Indicators currently published in the latest release. The informaton includes series information and 

Goal - Target hierarchy

POST /v1/sdg/DataAvailability/GetWorldbyGoal Returns data for Global Summary Of the World for the Mixed line and bar Chart

GET /v1/sdg/GeoArea/Tree Returns a tree of regions and countries

GET /v1/sdg/Goal/{goalCode}/Target/List
Returns the list of all Targets available for a Goal.

```python
url_goal_list= "https://unstats.un.org/sdgapi/v1/sdg/Goal/List"
global_goals_list = requests.get(url_goal_list)
global_goals_list = global_goals_list.json()
display(global_goals_list)
len(global_goals_list)
```

```python

def dataframe_from_list_dic(list):
    cols = list[0].keys
    dataframe1 = pd.DataFrame(columns=[cols])
    for i in range(0, len(list)):
        data = list[i]
        dataframe2 = pd.DataFrame(data, index=[0])  
        dataframe1 = pd.concat([dataframe1, dataframe2], ignore_index=True)
    dataframe1.drop(dataframe1.columns[0], axis=1, inplace = True)
    return dataframe1

global_goals = dataframe_from_list_dic(global_goals_list)


```

```python
global_goals.to_csv("global_goals.csv")
```

```python
geoarea_url= "https://unstats.un.org/sdgapi/v1/sdg/GeoArea/List"
geoarea_url_list = requests.get(geoarea_url)
geoarea_list = geoarea_url_list.json()
display(geoarea_list)
len(geoarea_list)
```

```python
geoarea_list = dataframe_from_list_dic(geoarea_list)
geoarea_list
```

```python
geoarea_list.to_csv("geoarea_list.csv", index=False)
```

```python
def datafram2(list):
    cols = list[0].keys
    dataframe1 = pd.DataFrame(columns=[cols])
    for i in range(0, min(19, len(list))):
        data = list[i]
        dataframe2 = pd.DataFrame.from_dict(data, orient='index').T
        # Ensure index is provided
        dataframe1 = pd.concat([dataframe1, dataframe2], ignore_index=True)
    dataframe1.drop(dataframe1.columns[0], axis=1, inplace=True)
    return dataframe1

indicators = datafram2(indic)
```
