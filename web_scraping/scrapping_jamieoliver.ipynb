```python


def get_list_recipes(components):
    recipes = []
    urls = []

    for component in components:
   
        url = component.find("a").get('href')
        urls.append(url)

        recipe = component.find("div", class_="recipe-title").text.strip()
        recipes.append(recipe)

    df = pd.DataFrame({
        "url": urls,
        "recipes": recipes,
    })

    return df


def get_df_from_url(url):
    sleep(5)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    components = soup.find_all('div', class_='recipe-block')
    df = get_list_recipes(components)
    return df



```

```python
# Lists to store extracted data
urls = []
recipes = []
ingredients_list = []
instructions = []
category = []
cuisine = []
diet = []
serves = []
time_prep = []
time_cook = []
nutrients_table = []

def get_recipe(components, url):
    for component in components:

        try:
            urls.append(url)
            recipe = component.find("h1").text.strip()
            recipes.append(recipe)

            ingredients = component.find("ul", class_="ingred-list").text.strip()
            ingredients_list.append(ingredients)

            instruc = component.find("div", class_="method-p").text.strip()
            instructions.append(instruc)

            #cat = component.find("span", class_="tasty-recipes-category").text.strip()
            category.append("undefined")

           # cuisines = component.find_all("span", class_="tasty-recipes-cuisine")
            #cuisines_text = ', '.join(cuisine.text.strip() for cuisine in cuisines)
            cuisine.append("undefined")

            time_p = component.find("div", class_="recipe-detail time").text.strip()
            time_prep.append(time_p)

            time_c = component.find("div", class_="recipe-detail time").text.strip()
            time_cook.append(time_c)

            d = component.find("div", class_="special-diets-wrapper").text.strip()[0]
            diet.append(d)

            y = component.find("div", class_="recipe-detail serves").text.strip()
            serves.append(y)

            recipe_nutrition = component.find_all("div", class_="inner")
            list_info = []
            for i in recipe_nutrition:
                list_info.append(i.text.strip())
            nutrients_table.append(list_info)

    
        except Exception as e:
            print(f'Error processing {url}: {e}')
   
            urls.append(url)
            recipes.append("error")
            ingredients_list.append("error")
            instructions.append("error")
            category.append("error")
            cuisine.append("error")
            time_prep.append("error")
            time_cook.append("error")
            diet.append("error")
            serves.append("error")
            nutrients_table.append("error")


def get_df_from_url_list(pages):
    for url in pages:
        print('Working on page' + str(url))
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        components = soup.find_all('div', class_='container recipe-container')
        try:
            sleep(5)
            get_recipe(components, url)
        except Exception as e:
            print(f'Error processing {url}: {e}')
    data = list(zip(urls, recipes, ingredients_list, instructions, category, cuisine, diet, serves, time_prep, time_cook, nutrients_table))
    df = pd.DataFrame(data, columns=["url", "recipes", "list_ingredients", "instructions", "category", "cuisine", "diet", "serves", "time_prep","time_cook", "nutrients_table"])
    return df

#pages = list_urls_veg
#final_df = get_df_from_url(pages)
```

```python

# FINAL FINAL FONCTION SOS

def process_urls_in_batches(url_list, batch_size=50):
    urls_tried = []
    urls_error = []
    batches = [url_list[i:i+batch_size] for i in range(0, len(url_list), batch_size)]

    for idx, batch in enumerate(batches):
        print(f"Processing batch {idx + 1}...")
        urls_tried.extend(batch)
        df_batch = pd.DataFrame(columns=["url", "recipes", "list_ingredients", "instructions", "category", "cuisine", "diet", "serves", "time_prep", "time_cook"])    
        
        try:
            df_processed = get_df_from_url_list(batch)
            display(df_processed)
            df_batch = pd.concat([df_batch, df_processed], ignore_index=True)
        except Exception as e:
            print(f'Error processing {url}: {e}')
            urls_error.append(url)


        csv_file = f'jamie_oliver_recipe_data_batch_{idx+1}.csv'
        df_batch.to_csv(csv_file, index=False)
        print(f"Data for batch {idx+1} saved to {csv_file}")

    print("Processing complete.")
    return urls_tried, urls_error



```
