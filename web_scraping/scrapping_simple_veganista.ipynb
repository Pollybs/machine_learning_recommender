```python


def get_list_recipes(components):
    recipes = []
    urls = []

    for component in components:
   
        url = component.find("a").get('href')
        urls.append(url)

        recipe = component.find("a").text.strip()
        recipes.append(recipe)

    df = pd.DataFrame({
        "url": urls,
        "recipes": recipes,
    })

    return df


def get_df_from_url(url):
    sleep(5)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    components = soup.find_all('h2', class_='entry-title')
    df = get_list_recipes(components)
    return df

#get_df_from_url('https://simple-veganista.com/recipes/all-recipes/')

```

### getting the nutrients table

```python
# Lists to store extracted data
urls = []
recipes = []
nutrients_table = []

def get_nutriments_table(components, url):
    for component in components:

        try:
            urls.append(url)
            recipe_nutrition = component.find_all("tr") #class_="label-table"
            list_tr = []
            for i in recipe_nutrition:

                list_tr.append(i.text.strip())
            nutrients_table.append(list_tr)

    
        except Exception as e:
            print(f'Error processing {url}: {e}')
   
            urls.append(url)
            nutrients_table.append("error")


def get_df_from_url_list(pages):
    for url in pages:
        print('Working on page' + str(url))
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        components = soup.find_all('table', class_='label-table')
        try:
            sleep(5)
            get_nutriments_table(components, url)
        except Exception as e:
            print(f'Error processing {url}: {e}')
    data = list(zip(urls, nutrients_table))
    df = pd.DataFrame(data, columns=["url", "nutrients_table"])
    return df

#pages = list_urls_veg
#final_df = get_df_from_url(pages)
```

```python
# Lists to store extracted data
urls = []
recipes = []
ingredients_list = []
instructions = []
category = []
cuisine = []
diet = []
serves = []
time_prep = []
time_cook = []

def get_recipe(components, url):
    for component in components:
        # Extract recipe name and link
        try:
            urls.append(url)
            recipe = component.find("h2").text.strip()
            recipes.append(recipe)

            ingredients = component.find("div", class_="tasty-recipes-ingredients-body").text.strip()
            ingredients_list.append(ingredients)

            instruc = component.find("div", class_="tasty-recipes-instructions-body").text.strip()
            instructions.append(instruc)

            cat = component.find("span", class_="tasty-recipes-category").text.strip()
            category.append(cat)

            cuisines = component.find_all("span", class_="tasty-recipes-cuisine")
            cuisines_text = ', '.join(cuisine.text.strip() for cuisine in cuisines)
            cuisine.append(cuisines_text)

            time_p = component.find("span", class_="tasty-recipes-prep-time").text.strip()
            time_prep.append(time_p)

            time_c = component.find("span", class_="tasty-recipes-cook-time").text.strip()
            time_cook.append(time_c)

            d = component.find("span", class_="tasty-recipes-diet").text.strip()
            diet.append(d)

            y = component.find("span", class_="tasty-recipes-yield").text.strip()
            serves.append(y)

    
        except Exception as e:
            print(f'Error processing {url}: {e}')
            # Append "error" to the respective column
            urls.append(url)
            recipes.append("error")
            ingredients_list.append("error")
            instructions.append("error")
            category.append("error")
            cuisine.append("error")
            time_prep.append("error")
            time_cook.append("error")
            diet.append("error")
            serves.append("error")


def get_df_from_url_list(pages):
    for url in pages:
        print('Working on page' + str(url))
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        components = soup.find_all('div', class_='tasty-recipes')
        try:
            sleep(5)
            get_recipe(components, url)
        except Exception as e:
            print(f'Error processing {url}: {e}')
    data = list(zip(urls, recipes, ingredients_list, instructions, category, cuisine, diet, serves, time_prep, time_cook))
    df = pd.DataFrame(data, columns=["url", "recipes", "list_ingredients", "instructions", "category", "cuisine", "diet", "serves", "time_prep","time_cook"])
    return df

#pages = list_urls_veg
#final_df = get_df_from_url(pages)
```
