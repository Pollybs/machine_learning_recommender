```python


website = """
########################################################
#              WEBSITE: https://veganuary.com/recipes/         #
########################################################
#"""
print(website)

urls = []
recipes = []
ingredients_list = []
instructions = []
diet = []
serves = []
time_prep = []
time_cook = []

tart_time = datetime.now()
print('Crawl starting time:', start_time.time())



reci_list = reci_li

driver = webdriver.Chrome()



urls = []
recipes = []
ingredients_list = []
instructions = []
diet = []
serves = []
time_prep = []
time_cook = []

for url in reci_list:
    try:
        print('Working on', url)
        driver.get(url)
        sleep(randint(3,6))  
        
    
        title_element = WebDriverWait(driver, 10).until(
            EC.visibility_of_element_located((By.XPATH, '//h1[@class="hero__title"]'))
        )
        title_text = title_element.text
        recipes.append(title_text)
        
    
        prep_time_element = driver.find_element(By.XPATH, '//div[@class="tooltip-single"][1]/div[@class="tooltip-text-recipe"]')
        prep_time = prep_time_element.text.split(': ')[1]
        time_prep.append(prep_time)

        cook_time_element = driver.find_element(By.XPATH, '//div[@class="tooltip-single"][2]/div[@class="tooltip-text-recipe"]')
        cook_time_e = cook_time_element.text.split(': ')[1]
        time_cook.append(cook_time_e)

    
        servings_element = driver.find_element(By.XPATH, '//div[@class="tooltip-single"][3]/div[@class="tooltip-text-recipe"]')
        servings = servings_element.text.split(': ')[1]  
        serves.append(servings)
    
        ingred_element1 = driver.find_element(By.XPATH, '//div[@class="recipe__ingredients"]')
        ingred_element = ingred_element1.text 
        ingredients_list.append(ingred_element)

        prep_element1 = driver.find_element(By.XPATH, '//div[@class="col-lg-8"]')
        prep_element = prep_element1.text 
        instructions.append(prep_element)


        diet.append('1')
        urls.append(url)

    except Exception as e:
        print(f'Recipe {url} does not exist or could not be processed. Error:', e)


data2 = list(zip(urls, recipes, ingredients_list, instructions, diet, serves, time_prep, time_cook))
df2 = pd.DataFrame(data2, columns=["url", "recipes", "list_ingredients", "instructions", "diet", "serves", "time_prep","time_cook" ])


driver.quit()


csv_file = 'veganuary_recipe_trytry.csv'
df2.to_csv(csv_file, index=False)

#print('Crawl finished. CSV file saved as:', csv_file)
#print('Crawl time:', datetime.now() - start_time)
```
